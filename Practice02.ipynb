```python
## visualization1

plt.figure(figsize=[5,3])
plt.hist(dataframe['Pregnancies'],color='g',edgecolor='black')
plt.plot()
plt.show()
```

```python
## visualization2

plt.figure(figsize=[5,3])
sns.scatterplot(x='Age',y='BloodPressure',data=dataframe,color='orange')
plt.xlabel('Age')
plt.ylabel('BloodPressure')
plt.title('Distribution of Age Vs BloodPressure')
plt.show()
```

```python
## visualization3
plt.figure(figsize=[5,3])
plt.hist(dataframe['Glucose'],color='g',edgecolor='black')
plt.xlabel('Frequecy')
plt.xlabel('Glucose')
plt.title('Distribution of Glucose')
plt.plot()
plt.show()
```
